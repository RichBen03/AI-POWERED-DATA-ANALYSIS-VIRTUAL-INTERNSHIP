```python
import pandas as pd
import numpy as np
```

```python
# reading the dataset
slu_data = pd.read_csv("SLU Opportunity Wise Data.csv")
```

```python
# check for the first 100 rows of the data set
slu_data.head()
```

```python
# check for any missing data in by columns
missing_data = slu_data.isnull().sum()
missing_data[:]
```

Missing data in the data set in the columns include:
1. The opportunity start date = 3794 counts
2. Institution Name = 5
3. Current/Intended Major = 5

```python
total_cells = np.product(slu_data.shape)
total_missing = missing_data.sum()
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```
